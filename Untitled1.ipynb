```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Superstore.csv')
display(df.head())
```

```python
df.info()
```

```python
plt.figure()
sns.barplot(x='Category', y='Amount', data=df)
plt.title('Category Wise Sales')
plt.show()
```

```python
display(df.describe())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Order Date'].dt.month)['Amount'].sum()

plt.figure()
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
region_sales = df.groupby('State')['Amount'].sum()

plt.figure()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%')
plt.title('State Wise Sales Distribution')
plt.show()
```
